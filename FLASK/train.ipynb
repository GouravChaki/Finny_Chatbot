```python
import numpy as np
import pandas as pd
from sklearn import datasets,linear_model,preprocessing,metrics
from sklearn.model_selection import train_test_split
from flask import Flask, request
```

```python
db=pd.read_csv('./temporary_store_data.csv')
db['date']=0
db.head() #apple stocks dataset present
```

```python
#user defined function to remove hyphen from date (Data- preprocessing)
for x in (db.index) :
    timestamp=db['Date'][x]
    t=''
    for x1 in timestamp :
        if(x1=='-'):
            continue
        elif (x1==' '):
            break
        else:
            t=t+x1
    db.loc[x,'date']=int(t)

# data-preprocessing
# over-fitting
```
