```python
import numpy as np
import pandas as pd
from sklearn import datasets,linear_model,preprocessing,metrics
from sklearn.model_selection import train_test_split
```

```python
db=pd.read_csv('intraday_5min_IBM.csv')

db['date']=0
db
```

```python
# to make price column
for x in (db.index) :
    timestamp=db['timestamp'][x]
    t=''
    for x1 in timestamp :
        if(x1=='-'):
            continue
        else:
            t=t+x1
    db.loc[x,'date']=int(t)
db
```

```python
X=db[['date']]
X #label --- price
```

```python
Y=db[['open','high','low','close']]
Y #features
```
